```python
def project_to_2d(cubes, dim1, dim2):
    """
    Project high-dimensional BAdicCubes to a 2D plane.
    :param cubes: An array of BAdicCubes in high dimensions.
    :param dim1: Index of the first dimension to project.
    :param dim2: Index of the second dimension to project.
    :return: 2D representation of cubes.
    """
    projected_cubes = []
    for cube in cubes:
        b_adic_ranges = cube.b_adic_ranges
        projected_ranges = [b_adic_ranges[dim1], b_adic_ranges[dim2]]
        projected_cubes.append(BAdicCube(projected_ranges, cube.level, cube.index))
    plot_b_adic_cubes(projected_cubes)
    return np.asarray(projected_cubes)
```
